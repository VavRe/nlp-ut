```python
import hazm as hz
import nltk 
import chardet
```

```python
DATASET_PATH = "./data/hp_fa.txt"
```

```python
with open(DATASET_PATH, 'rb') as f:
    result = chardet.detect(f.read())

print(result['encoding'])
```

```python
with open(DATASET_PATH, "r", encoding="UTF-8-SIG") as f:
    hp_fa = f.read() 
print(hp_fa)
```

```python
normalizer = hz.Normalizer()
hp_fa_norm = normalizer.normalize(hp_fa)
print(hp_fa_norm)
```

```python
## Adding <s> and </s> tokens
sent_tokenizer = hz.SentenceTokenizer()
hp_fa_norm_padded = ""
sentences = sent_tokenizer.tokenize(hp_fa_norm)
for sentence in sentences:
    hp_fa_norm_padded += f"<s> {sentence} </s> \n"
print(hp_fa_norm_padded)
```

```python
hp_fa_norm_tokenized = hz.word_tokenize(hp_fa_norm_padded) 
print(hp_fa_norm_tokenized)
```

```python
import pickle

with open("./hp_fa_norm_tokenized_arr.pickle", "wb") as f:
    pickle.dump(hp_fa_norm_tokenized,f)
```
