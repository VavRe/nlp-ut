```python
import nltk
```

```python
import pickle 

with open("./hp_fa_norm_tokenized_arr.pickle" , "rb") as f:
    hp_fa = pickle.load(f)

print(hp_fa)
```

```python
import nltk.lm.preprocessing

bigrams = list(nltk.ngrams(hp_fa, 2))
bigram_dist = nltk.probability.ConditionalFreqDist(bigrams)
bigram_language_model = nltk.probability.ConditionalProbDist(bigram_dist,nltk.probability.LaplaceProbDist, bins=len(bigram_dist))
```

```python
import random
def generate_samples(language_model, min_token_len=12,max_token_len=24,sencences_count=5):
    sentences_count = 5 
    min_token_len = 12
    max_token_len = 24
    start_token = "<s>"
    sentences = list([])
    for i in range (sentences_count):
        sentences.append([start_token])
        sentence_length = random.randint(min_token_len,max_token_len)
        for j in range(sentence_length):
            last_token = sentences[i][-1]
            token = language_model[last_token].generate()
            sentences[i].append(token)
    return sentences
```

```python
def show_samples(sentences):
    for i in range(len(sentences)):        
        sentences[i][0] = u'\u202B'
        print(f"{i+1}. {' '.join(sentences[i][1:])} \n")
```

```python
bigram_sentences = generate_samples(bigram_language_model)
show_samples(bigram_sentences)
```
