```python
import nltk.lm.preprocessing

def make_ngram_lm(n_):
    ngrams = list(nltk.ngrams(hp_fa, n_))
    ngram_dist = nltk.probability.ConditionalFreqDist(ngrams)
    ngram_language_model = nltk.probability.ConditionalProbDist(ngram_dist,nltk.probability.LaplaceProbDist, bins=len(ngram_dist))
    return ngram_language_model
```

```python
bigram_lm = make_ngram_lm(2)
```

```python
import random
def generate_samples(language_model, min_token_len=12,max_token_len=24,sencences_count=5):
    sentences_count = 5 
    min_token_len = 12
    max_token_len = 24
    start_token = "<s>"
    sentences = list([])
    for i in range (sentences_count):
        sentences.append([start_token])
        sentence_length = random.randint(min_token_len,max_token_len)
        for j in range(sentence_length):
            last_token = sentences[i][-1]
            token = language_model[last_token].generate()
            sentences[i].append(token)
    return sentences
```
